```python
from sklearn import datasets, metrics
import tensorflow as tf
import numpy as np
from sklearn.cross_validation import train_test_split
%matplotlib inline
```

```python
# Data set of IRIS
iris = datasets.load_iris()

# Load datasets and split them for training and testing
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target)


####### TENSORFLOW #####

# Here is tensorflow's syntac for defining features.
# We must specify that all features have real-value data
feature_columns = [tf.contrib.layers.real_valued_column("", dimension=4)]

optimizer = tf.train.GradientDescentOptimizer(learning_rate=.1)


# Build a linear classifier (logistic regression)
# note we have to tell tensorfow the number of classes we are looking for
classifier = tf.contrib.learn.LinearClassifier(feature_columns=feature_columns,
                                               optimizer=optimizer,
                                            n_classes=3)

# Fit model. Uses error optimization techniques like stochastic gradient descent
classifier.fit(x=X_train,
               y=y_train,
               steps=1000)  # number of iterations

# Evaluate accuracy.
accuracy_score = classifier.evaluate(x=X_test,
                                     y=y_test)["accuracy"]
print('Accuracy: {0:f}'.format(accuracy_score))

# Classify two new flower samples.
new_samples = np.array(
    [[6.4, 3.2, 4.5, 1.5], [5.8, 3.1, 5.0, 1.7]], dtype=float)
y = classifier.predict(new_samples)
print('Predictions: {}'.format(str(y)))
```

```python
from glob import glob  # file reading package
from pylab import imread, imsave, subplot, imshow, title, gray, figure, show, NullLocator
from numpy import mean, cov, dot, linalg, size, argsort
import PIL  #python's imaging library
from PIL import Image
```

```python
# read in the images

image_dict = {'image':[], 'target': []}
code_to_target = {}
i = 0
size = 128, 128
for folder in glob('../data/objects/*'):
    name_of_object= folder.split('/')[-1]
    if name_of_object not in ['Motorbikes', 'octopus', 'accordion', 'stop_sign']:
        continue
    code_to_target[i] = name_of_object
    print name_of_object
    for f in glob(folder+'/*')[:20]:  # only take the first 50 images
        try:
            img = Image.open(f)
            img = img.resize(size, Image.ANTIALIAS) # resize image to 128 by 128 for ease
            img.save(f, "JPEG") # resave the image back to the folder
            row = imread(f) # load an image  # load it into numpy
            row = row.reshape(49152)  # turn it into a 1d array
            image_dict['image'].append(row)
            image_dict['target'].append(i)
        except Exception as e:
            pass
        
    i += 1
```

```python
image_dict['image'][0].shape
```

```python
images_X = np.array(image_dict['image']).astype(float)
images_y = np.array(image_dict['target'])
```

```python
# Load datasets.
X_train, X_test, y_train, y_test = train_test_split(images_X, images_y)
```

```python
## WARNING SLOWWWWW ##

# Specify that all features have real-value data
feature_columns = [tf.contrib.layers.real_valued_column("", dimension=49152)]

optimizer = tf.train.GradientDescentOptimizer(learning_rate=.01)


# Build DNN
classifier = tf.contrib.learn.DNNClassifier(feature_columns=feature_columns,
                                            hidden_units=[100,100,100,100],
                                            optimizer=optimizer,
                                            n_classes=len(code_to_target))

# Fit model.
classifier.fit(x=X_train,
               y=y_train,
               steps=10000)
```

```python
# Evaluate accuracy.
accuracy_score = classifier.evaluate(x=X_test,
                                     y=y_test)["accuracy"]
print('Accuracy: {0:f}'.format(accuracy_score))
```
